```python
import re
import numpy as np
def sayilari_bul(metin):
    if pd.isnull(metin):
        return np.nan
    # Sayıları bul
    sayilar = re.findall(r'\d+', metin)
    # Sayıları integer'a çevir ve birleştir
    sayilar = ''.join(sayilar)
    if sayilar: # Eğer sayılar boş değilse integer'a çevir
        return int(sayilar)
    return np.nan

def harfleri_bul(metin):
    if pd.isnull(metin):
        return np.nan
    # Harfleri bul ve birleştir
    harfler = re.findall(r'[a-zA-Z]+', metin)
    harfler = ''.join(harfler)
    return harfler if harfler else np.nan

# Örnek pandas serisi
seri = df["Invoice"].astype(str)

# Sayıları ve harfleri ayrı ayrı bul
sayilar_seri = seri.apply(sayilari_bul)
harfler_seri = seri.apply(harfleri_bul)

# Sonuçları bir DataFrame'de birleştir
a = pd.DataFrame({
    'Sayılar': sayilar_seri,
    'Harfler': harfler_seri
})
a
```
