```python
def content_based_recommender(title,cosine_sim,df):
    #indexleri oluşturma
    indices = pd.Series(df.index,index =df["title"])
    indices = indices[~indices.index.duplicated(keep="last")]
    #title indexini yakalama
    movie_index = indices[title]
    #title a göre benzerlik skorları oluştur
    similarity_scores = pd.DataFrame(cosine_sim[movie_index],columns=["score"])
    movie_indices = list(similarity_scores.sort_values(by="score",ascending=False)[1:11].index)
    return df["title"].iloc[movie_indices]

content_based_recommender("Sherlock Holmes",cosine_sim,df)
```

```python
def calculate_cosine_sim(df):
    tfidf = TfidfVectorizer(stop_words="english")
    df["overview"] = df["overview"].fillna("")
    tfidf_matrix = tfidf.fit_transform(df["overview"])
    cosine_sim = cosine_similarity(tfidf_matrix,tfidf_matrix)
    return cosine_sim
```

```python
cosine_sim = calculate_cosine_sim(df)
```

```python
content_based_recommender("The Dark Knight Rises",cosine_sim,df)
```
