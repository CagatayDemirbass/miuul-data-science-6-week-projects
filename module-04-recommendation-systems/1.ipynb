```python
def retail_data_prep(df):
    df.dropna(inplace=True)
    df = df.loc[~df["Invoice"].str.contains("C",na=False)]
    df = df.loc[df["Quantity"]>0]
    df = df.loc[df["Price"]>0]
    return df

df = retail_data_prep(df)
df.describe()
```

```python
def outlier_thresh(df,variable):
    q1 = df[variable].quantile(0.01)
    q3 = df[variable].quantile(0.99)
    iqr = q3 - q1
    upper_lim = q3 + 1.5*iqr
    lower_lim = q1 - 1.5*iqr
    df[variable].clip(upper=upper_lim,inplace=True)

outlier_thresh(df,"Quantity")
outlier_thresh(df,"Price")
df.describe()
```

```python
frequent_itemsets = apriori(fr_inv_pro_df,min_support=0.01,use_colnames=True)
frequent_itemsets.sort_values("support",ascending=False)
```

```python
# antecedent support: x / tüm işlem sayısı. x in tüm alışveriş içinde geçme yüzdesi
# consequent support : y / tüm işlem sayısı. y nin tüm alışveriş içinde geçme yüzdesi
# support x+y / tüm işlem sayısı. x ve y nin tüm alışveriş içinde birlikte geçme yüzdesi
# confidence x alındığında y nin satılma olasılığı
# x satın alındığında y nin satılma olasılığı lift kat kadar artar 1 in üstünde ise artar altında ise azalır
```

```python
rules=association_rules(frequent_itemsets,metric="support",min_threshold=0.01)
rules.head()
```

```python
rules.loc[(rules["support"]>0.05) & (rules["confidence"]>0.1) & (rules["lift"]>5)].sort_values("confidence",ascending=False)
```

Evet, doğru. support değeri, belirli bir öğe setinin (örneğin, iki ürünün bir arada görülmesi gibi) tüm işlemler arasındaki oranını belirtir.

Örneğin, bir market alışveriş veri setinde eğer elma ve muzun birlikte satın alındığı 50 işlem varsa ve toplamda 1000 işlem gerçekleştiyse, "elma ve muz birlikte satın alınır" kuralının support değeri 50/1000, yani 0.05 olacaktır. Bu, elma ve muzun birlikte satın alındığı işlemlerin tüm işlemlerin %5'ini oluşturduğu anlamına gelir.

Bu bilgi, öğelerin birlikte satın alınma sıklığını anlamamıza yardımcı olur ve bu tür bilgiler genellikle ürün yerleştirme, indirimler, ürün önerileri ve diğer pazarlama stratejileri oluşturmak için kullanılır.

confidence değeri, bir ilişkili öğe seti kuralının güvenilirliğini ölçer. Yani, belirli bir öğe seti A'nın alındığını bildiğimizde, öğe seti B'nin de alınma olasılığını ifade eder.

Örneğin, eğer "elma ve muz" ile "süt" arasında bir ilişkili öğe kuralımız varsa, bu kuralın confidence değeri, elma ve muz alındığında sütün de alınma olasılığını gösterir.

Formal olarak, confidence(A -> B) = support(A, B) / support(A) şeklinde ifade edilir. Yani, A ve B birlikte alınma olasılığı (destek) / A'nın alınma olasılığı (destek).

Confidence değeri 0 ile 1 arasında bir değerdir. 1'e yaklaşan bir confidence değeri, A alındığında B'nin de alınma olasılığının yüksek olduğunu gösterir. Diğer bir deyişle, A'nın alınması B'nin alınmasını tahmin etmek için güçlü bir gösterge olabilir.


Lift, birliktelik kurallarında bir ölçüm metriği olarak kullanılır. Lift değeri, bir öğe seti A'nın varlığında öğe seti B'nin görülme olasılığının, B'nin genel görülme olasılığına göre ne kadar arttığını (veya azaldığını) gösterir.

Matematiksel olarak, lift(A -> B) = confidence(A -> B) / support(B) şeklinde hesaplanır.

Lift değeri 1'den büyükse, bu A'nın varlığının B'nin görülme olasılığını artırdığını gösterir, yani A ve B birbirine bağımlıdır. Lift değeri 1 ise, bu A ve B'nin birbirinden bağımsız olduğunu gösterir, yani A'nın varlığı B'nin görülme olasılığını etkilemez. Lift değeri 1'den küçükse, bu A'nın varlığının B'nin görülme olasılığını azalttığını gösterir.

Örneğin, elma ve muzun birlikte satın alınması ile sütün satın alınması arasında bir kuralımız varsa ve bu kuralın lift değeri 1.5 ise, bu elma ve muzun satın alındığı işlemlerde sütün satın alınma olasılığının, genel olarak sütün satın alınma olasılığına göre 1.5 kat daha fazla olduğunu gösterir. Bu, belirli bir pazarlama kampanyasının etkisini değerlendirmek ya da ürün yerleştirme gibi konuları düşünürken kullanışlı olabilir.

Evet, confidence değeri 1 olduğunda, bu "A alındığında B'nin de kesinlikle alındığı" anlamına gelir. Yani, veri setindeki tüm işlemlerde, her seferinde A öğesinin alındığı yerde B öğesi de alınmıştır.

Ancak, unutmayın ki bu durum geçmiş verilere dayalıdır ve gelecek alışverişler için mutlak bir garanti sağlamaz. Ayrıca, bu durumun sebebi başka faktörlere bağlı olabilir ve bu nedenle her zaman dikkatli bir şekilde analiz edilmelidir.

Son olarak, bir confidence değerinin 1 olması her zaman ideal bir durum değildir. Çok yüksek confidence değerleri, genellikle çok düşük support değerleri ile birlikte görülür. Yani, bu tür kurallar yalnızca veri setinin çok küçük bir bölümüne uygulanabilir, bu nedenle genellenebilirlikleri sınırlıdır.

Birliktelik kurallarının değerlendirilmesinde confidence yanında support ve lift gibi diğer metriklerin de dikkate alınması önemlidir.

```python
def create_rules(df,id=True,country="France"):
    df = df.loc[df["Country"]==country]
    df = create_matrix_df(df,id)
    frequent_itemsets = apriori(df,min_support=0.01,use_colnames=True)
    rules = association_rules(frequent_itemsets,metric="support",min_threshold=0.01)
    return rules

rules = create_rules(df)
rules
```

```python
def recommendation(rules_df,product_id,rec_count=1):
    sorted_rules = rules_df.sort_values("lift",ascending=False)
    recommendation_list = []
    for i,product in enumerate(sorted_rules["antecedents"]):
        for j in list(product):
            if j == product_id:
                recommendation_list.append(list(sorted_rules.iloc[i]["consequents"]))
                
    return recommendation_list[0:rec_count]

recommendation(rules,22727,rec_count=1)
```

```python
def recommendation(rules_df, product_id, rec_count=2):
    sorted_rules = rules_df.sort_values("lift", ascending=False)
    recommendation_list = []
    counter = 0

    for i, product in enumerate(sorted_rules["antecedents"]):
        if counter < rec_count:  # Continue as long as counter is less than rec_count
            for j in list(product):
                if j == product_id:
                    recommendation_list.append(list(sorted_rules.iloc[i]["consequents"]))
                    counter += 1  # Increase the counter each time we add a recommendation
        else:
            break  # If counter equals to rec_count, stop the loop
            
    return recommendation_list

recommendation_list= recommendation(rules,22727,rec_count=1)
```

```python
def recommendation(rules_df, product_id, rec_count=2):
    sorted_rules = rules_df.sort_values("lift", ascending=False)
    recommendation_list = []
    counter = 0
    stop_loop = False

    for i, product in enumerate(sorted_rules["antecedents"]):
        if stop_loop:
            break
            
        for j in list(product):
            if j == product_id:
                recommendation_list.append(list(sorted_rules.iloc[i]["consequents"]))
                counter += 1

                if counter >= rec_count:
                    stop_loop = True
                    break
                    
    return recommendation_list
recommendation_list= recommendation(rules,22727,rec_count=1)
```
