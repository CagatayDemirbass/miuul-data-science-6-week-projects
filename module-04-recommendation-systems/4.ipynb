```python
import pandas as pd
def create_user_movie_df():
    movie = pd.read_csv("movies.csv")
    rating = pd.read_csv("ratings.csv")
    df = pd.merge(movie, rating,on="movieId",how="left")
    movie_count = df["title"].value_counts()
    common_movies = list(movie_count[movie_count>1000].index)
    common_movies_df = df.loc[df["title"].isin(common_movies)]
    # common_users = common_movies_df["userId"].value_counts()
    # common_users = list(common_users[common_users>250].index)
    # common_movies_df = common_movies_df.loc[common_movies_df["userId"].isin(common_users)]
    user_movie_df = common_movies_df.pivot_table(index="userId",columns="title",values="rating")
    return user_movie_df

user_movie_df=create_user_movie_df()
user_movie_df
```

```python
def user_based_reccomender(random_user,user_movie_df,cor_th=.65,ratio=.6):
    movies_watched=list(user_movie_df.loc[random_user].dropna().index)
    user_count=user_movie_df[movies_watched].count(axis=1)
    perc = len(movies_watched) * ratio
    user_ids = list(user_count[user_count>perc].index)
    movies_watched_df = user_movie_df[movies_watched]
    final_df=movies_watched_df.loc[user_ids]
    corr_df= final_df.T.corr().unstack().sort_values()
    corr_df = pd.DataFrame(corr_df,columns=["corr"])
    corr_df.index.names = ["user_id_1","user_id_2"]
    corr_df = corr_df.reset_index()
    top_users=corr_df.loc[(corr_df["user_id_1"]==random_user) & (corr_df["corr"]>cor_th) & (corr_df["user_id_2"]!=random_user),["user_id_2","corr"]]
    top_users.rename(columns={"user_id_2":"userId"},inplace=True)
    rating = pd.read_csv("ratings.csv")
    top_users_rating=pd.merge(rating[["userId","movieId","rating"]],top_users,how="inner",on="userId")
    top_users_rating["weighted_rating"] = top_users_rating["corr"] * top_users_rating["rating"]
    recommendation_df=top_users_rating.groupby("movieId").agg({"weighted_rating":"mean"})
    recommendation_df.reset_index(inplace=True)
    recommentation_dict = {}
    recommentation_dict[random_user] =list(recommendation_df.nlargest(10,"weighted_rating")["movieId"].values)
    return recommentation_dict
```

```python
user_based_reccomender(random_user,user_movie_df,cor_th=.7)
```
