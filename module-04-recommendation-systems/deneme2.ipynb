```python
import pandas as pd
def create_user_movie_df():
    movie = pd.read_csv("movies.csv")
    rating = pd.read_csv("ratings.csv")
    df = pd.merge(movie, rating,on="movieId",how="left")
    movie_count = df["title"].value_counts()
    common_movies = list(movie_count[movie_count>1000].index)
    common_movies_df = df.loc[df["title"].isin(common_movies)]
    common_users = common_movies_df["userId"].value_counts()
    common_users = list(common_users[common_users>250].index)
    common_movies_df = common_movies_df.loc[common_movies_df["userId"].isin(common_users)]
    user_movie_df = common_movies_df.pivot_table(index="userId",columns="title",values="rating")
    return user_movie_df

user_movie_df=create_user_movie_df()
user_movie_df
```

```python
random_user=float(pd.Series(user_movie_df.index).sample(1,random_state=45).values)
random_user
```

```python
random_user_df = user_movie_df[user_movie_df.index==random_user]
movies_watched= random_user_df.columns[random_user_df.notna().any()].tolist()
movies_watched
```

```python
movies_watched_df=user_movie_df[movies_watched]
movies_watched_df
```

```python
user_movie_count=movies_watched_df.count(axis=1)
user_movie_count
```

```python
user_movie_count = user_movie_count.reset_index()
user_movie_count.columns = ["userId","movie_count"]
user_movie_count
```

```python
users_same_movies=user_movie_count.loc[user_movie_count["movie_count"]>200,"userId"]
users_same_movies
```

```python
final_df=pd.concat([movies_watched_df[movies_watched_df.index.isin(users_same_movies)],random_user_df[movies_watched]])
final_df
```

Pandas'ın drop_duplicates() fonksiyonu, yinelenen değerleri çıkarırken eşleşen değerlerin tamamına bakar. Ancak, bu kod parçacığı final_df.T.corr().unstack().sort_values().drop_duplicates() korelasyon değerlerini içeren bir seri oluşturduğu için, drop_duplicates() sadece korelasyon değerlerine bakar.

Özellikle, (id_1: 1234, id_2: 5678, corr: 0.8) ve (id_1: 99126, id_2: 817321, corr: 0.8) çiftleri, korelasyon değerlerine bakıldığında aynıdır (her ikisi de 0.8). Dolayısıyla, drop_duplicates() çağrısı yalnızca birini korur.

Eğer drop_duplicates() çağrısını bir dataframe üzerinde yapsaydınız ve bu dataframe ID bilgilerini içerseydi, o zaman drop_duplicates() ID'ler ve korelasyon değerlerinin bir kombinasyonuna bakardı. Bu durumda, (1234, 5678, 0.8) ve (99126, 817321, 0.8) birbirinden farklı olduğu için her ikisi de korunurdu.

Sonuç olarak, drop_duplicates() fonksiyonunun çıktısı, üzerinde çalıştığı veri yapısına bağlıdır. Bir pandas Serisi üzerinde çalıştığında, yalnızca Seri içindeki değerlere bakar ve bu değerler üzerinde yinelenenleri çıkarır. Ancak, bir DataFrame üzerinde çalıştığında, yinelenen satırları (yani tüm sütunlardaki değerlerin aynı olduğu satırları) çıkarır.
