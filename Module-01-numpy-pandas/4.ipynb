```python
# verinin sÃ¼tun isimleri ve bunlarÄ±n karÅŸÄ±lÄ±k geleceÄŸi grafik baÅŸlÄ±klarÄ±
fig = make_subplots(rows=5, cols=3)

for i, col in enumerate(df.columns):
    row = i // 3 + 1
    col =  i % 3 + 1
    fig.add_trace(go.Histogram(x=df[df.columns[i]]), row=row, col=col)

fig.update_layout(height=700, width=1100, title_text="Multiple Subplots",template="plotly_dark")
fig.show()
```
