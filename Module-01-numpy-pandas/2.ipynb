```python
import matplotlib.pyplot as plt

# Veriyi oluşturma
data = np.random.normal(10,4,(3,4))

# Veriyi tek boyutlu bir dizi haline getirme
flattened_data = data.flatten()
    
# Histogramı çizdirme
plt.hist(flattened_data, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histogram of Randomly Generated Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
