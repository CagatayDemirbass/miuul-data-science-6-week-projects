```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split,cross_val_score
df = pd.read_csv("advertising.csv")
df.head()
```

```python
X = df[["TV"]]
y = df["sales"]
lr = LinearRegression().fit(X,y)
lr.coef_[0] #[0][0] w
```

```python
lr.intercept_ #b
```

```python
lr.intercept_ +	230.1*lr.coef_[0]
```

```python
rd = Ridge(alpha=10).fit(X,y)
rd.intercept_
```

```python
rd.coef_[0] 
```

```python
rd.intercept_ +	230.1*rd.coef_[0]
```

```python
sns.regplot(x=X,y=y,scatter_kws={"color":"b","s":9},ci=False,color="r")
```

```python
y_pred=lr.predict(X)
mean_squared_error(y,y_pred)
```

```python
np.sqrt(mean_squared_error(y,y_pred))
```

```python
mean_absolute_error(y,y_pred)
```

```python
y.mean()
```

```python
y.std()
```

```python
lr.score(X,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor().fit(X,y)
rf.score(X,y)
```

```python
y_pred2= rf.predict(X)
mean_squared_error(y,y_pred2)
```

```python
np.sqrt(mean_squared_error(y,y_pred2))
```

```python
from xgboost import XGBRegressor
xg = XGBRegressor().fit(X,y)
y_pred3 = xg.predict(X)
mean_squared_error(y,y_pred3)
```

```python
np.sqrt(mean_squared_error(y,y_pred3))
```

```python
from xgboost import XGBRegressor
xg = XGBRegressor(booster="gblinear").fit(X,y)
y_pred3 = xg.predict(X)
mean_squared_error(y,y_pred3)
```
