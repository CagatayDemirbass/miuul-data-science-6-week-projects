```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
%matplotlib inline
import seaborn as sns
pd.set_option("display.max_columns",None)
df = sns.load_dataset("titanic")
df.head()
```

```python
df["sex"].value_counts().plot.bar(color = "orange")
```

```python
df["sex"].value_counts().plot(kind="bar",color="orange")
```

```python
df["age"].hist(color="orange",bins=50)
plt.show()
```

```python
plt.boxplot(df["fare"])
plt.show()
```

```python
x=np.array([10,170,26,5])
y=np.array([1,30,150,40])
plt.subplot(1,2,1)
plt.title("1")
plt.plot(x,marker="P",linestyle="dashed",color="r")
plt.subplot(1,2,2)
plt.title("2")
plt.plot(y,marker="D",color="orange",linestyle="dashdot")
plt.show()
```

```python
x=np.array([10,170,26,5])
y=np.array([1,30,150,40])
plt.subplot(1,3,1)
plt.title("1")
plt.plot(x,color='red')
plt.subplot(1,3,2)
plt.title("2")
plt.plot(y,color='green')
plt.subplot(1,3,3)
plt.title("3")
plt.plot(x,y,color='orange')
plt.show()
```
